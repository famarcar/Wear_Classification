```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
pd.__version__
```

```python
list_heads = ['desc']
list_folds = range(2, 50)
for l in list_folds:
    list_heads.append('fold_'+ str(l))
    list_heads.append('mean_'+ str(l))    
    list_heads.append('std_'+ str(l))

list_heads.append('blank')    
print(list_heads)
print(type(list_heads))
print(len(list_heads))
print(list_folds)
```

```python
Gaus_pd = pd.read_csv('GaussianNB1.txt', sep=" ", header=None) #, names=list_heads)
RaF_pd = pd.read_csv('RandomForest1.txt', sep=" ", header=None) #, names=list_heads)
SVC_pd = pd.read_csv('SVC1.txt', sep=" ", header=None) #, names=list_heads)
Tree_pd = pd.read_csv('DecisionTree1.txt', sep=" ", header=None) #, names=list_heads)

Gaus_pd.columns = list_heads
RaF_pd.columns = list_heads
SVC_pd.columns = list_heads
Tree_pd.columns = list_heads

del Gaus_pd['blank']
del RaF_pd['blank']
del SVC_pd['blank']
del Tree_pd['blank']
```

```python
import numpy as np
import matplotlib.pyplot as plt
#%matplotlib inline

list_folds = range(2, 50)

list_heads =[]
for l in list_folds:
    list_heads.append('mean_'+ str(l)) 

    
only_means_pd = RaF_pd.filter(items=list_heads)
#print(only_means_pd)

print("list folds: ", len(list_folds))
print("list folds: ", len(list_folds[:-1]))
print("only_means_pd: ", only_means_pd.shape)

label_exp =[]
mean_exp = []
max_exp = []

for l in range(len(list_folds)):
    only_means_pd.loc[l].plot( alpha=0.3) #, label="a")
    print("exp: ", RaF_pd["fold_2"][l], " mean: ", round(np.mean(only_means_pd.loc[l]), 3), " max: ",  round(np.max(only_means_pd.loc[l]), 3))
    label_exp.append(RaF_pd["fold_2"][l])
    mean_exp.append(round(np.mean(only_means_pd.loc[l]), 3))
    max_exp.append(round(np.max(only_means_pd.loc[l]), 3))
    

plt.xlim([0,50])
plt.ylim([0, 1])
#plt.legend()
type(np.mean(only_means_pd.loc[0]))
```

```python
plt.plot(label_exp, mean_exp)
plt.plot(label_exp, max_exp)
plt.xlabel('x axis label')
plt.ylabel('y axis label')
plt.title('experiment summary')
plt.legend(['mean', 'max'])
plt.show()
```

```python
import heapq
heapq.nlargest(3, mean_exp)
print(type(heapq.nlargest(3, mean_exp)))
mean_exp.index(heapq.nlargest(1, mean_exp))
```

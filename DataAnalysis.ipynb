```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
pd.__version__
```

creating head names to put into each read file

```python
list_heads = ['desc']
list_folds = range(2, 50)
for l in list_folds:
    list_heads.append('fold_'+ str(l))
    list_heads.append('mean_'+ str(l))    
    list_heads.append('std_'+ str(l))

list_heads.append('blank')    
print(list_heads)
print(type(list_heads))
print(len(list_heads))
print(list_folds)
```

```python
Gaus_pd = pd.read_csv('GaussianNB1.txt', sep=" ", header=None) #, names=list_heads)
RaF_pd = pd.read_csv('RandomForest1.txt', sep=" ", header=None) #, names=list_heads)
SVC_pd = pd.read_csv('SVC1.txt', sep=" ", header=None) #, names=list_heads)
Tree_pd = pd.read_csv('DecisionTree1.txt', sep=" ", header=None) #, names=list_heads)

Gaus_pd.columns = list_heads
RaF_pd.columns = list_heads
SVC_pd.columns = list_heads
Tree_pd.columns = list_heads

del Gaus_pd['blank']
del RaF_pd['blank']
del SVC_pd['blank']
del Tree_pd['blank']
```

```python
import numpy as np
import matplotlib.pyplot as plt
#%matplotlib inline

list_folds = range(2, 50)

list_means =[]
list_std =[]
for l in list_folds:
    list_means.append('mean_'+ str(l)) 
    list_std.append('std_'+ str(l)) 



only_means_Gaus_pd = Gaus_pd.filter(items=list_means)
only_means_RaF_pd = RaF_pd.filter(items=list_means)
only_means_SVC_pd = SVC_pd.filter(items=list_means)
only_means_Tree_pd = Tree_pd.filter(items=list_means)


only_std_Gaus_pd = Gaus_pd.filter(items=list_std)
only_std_RaF_pd = RaF_pd.filter(items=list_std)
only_std_SVC_pd = SVC_pd.filter(items=list_std)
only_std_Tree_pd = Tree_pd.filter(items=list_std)

print(type(only_means_RaF_pd))
#print((only_means_pd))

print("list folds: ", len(list_folds))
print("list folds: ", len(list_folds[:-1]))
print("only_means_pd: ", only_means_Gaus_pd.shape)

label_exp =[]

means_Gaus_pd = []
means_RaF_pd = []
means_SVC_pd = []
means_Tree_pd = []

max_Gaus_pd = []
max_RaF_pd = []
max_SVC_pd = []
max_Tree_pd = []

std_Gaus_pd = []
std_RaF_pd = []
std_SVC_pd = []
std_Tree_pd = []

sem_Gaus_pd = []
sem_RaF_pd = []
sem_SVC_pd = []
sem_Tree_pd = []


#uno todos los archivos...

print("range: ", range(len(list_folds)))
for l in range(len(list_folds)):
    #only_means_pd.loc[l].plot( alpha=0.3) #, label="a")
    #print("exp: ", RaF_pd["fold_2"][l], " mean: ", round(np.mean(only_means_pd.loc[l]), 3), " max: ",  round(np.max(only_means_pd.loc[l]), 3))
    label_exp.append(RaF_pd["fold_2"][l])
    
    means_Gaus_pd.append(round(np.mean(only_means_Gaus_pd.loc[l]), 3))
    means_RaF_pd.append(round(np.mean(only_means_RaF_pd.loc[l]), 3))
    means_SVC_pd.append(round(np.mean(only_means_SVC_pd.loc[l]), 3))
    means_Tree_pd.append(round(np.mean(only_means_Tree_pd.loc[l]), 3))

    std_Gaus_pd.append(round(np.std(only_means_Gaus_pd.loc[l]), 3))
    std_RaF_pd.append(round(np.std(only_means_RaF_pd.loc[l]), 3))
    std_SVC_pd.append(round(np.std(only_means_SVC_pd.loc[l]), 3))
    std_Tree_pd.append(round(np.std(only_means_Tree_pd.loc[l]), 3))

    sem_Gaus_pd.append(round(np.std(only_means_Gaus_pd.loc[l]/np.sqrt(len(only_means_Gaus_pd.loc[l]))), 3))
    sem_RaF_pd.append(round(np.std(only_means_RaF_pd.loc[l]/np.sqrt(len(only_means_Gaus_pd.loc[l]))), 3))
    sem_SVC_pd.append(round(np.std(only_means_SVC_pd.loc[l]/np.sqrt(len(only_means_Gaus_pd.loc[l]))), 3))
    sem_Tree_pd.append(round(np.std(only_means_Tree_pd.loc[l]/np.sqrt(len(only_means_Gaus_pd.loc[l]))), 3))

    max_Gaus_pd.append(round(np.max(only_means_Gaus_pd.loc[l]), 3))
    max_RaF_pd.append(round(np.max(only_means_RaF_pd.loc[l]), 3)) 
    max_SVC_pd.append(round(np.max(only_means_SVC_pd.loc[l]), 3)) 
    max_Tree_pd.append(round(np.max(only_means_Tree_pd.loc[l]), 3))
    


type(np.mean(only_means_Gaus_pd.loc[0]))
```

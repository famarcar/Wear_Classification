# Compile, visualize and analyze results

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
pd.__version__
```

```python
list_heads = ['desc']
list_folds = range(2, 50)
for l in list_folds:
    list_heads.append('fold_'+ str(l))
    list_heads.append('mean_'+ str(l))    
    list_heads.append('std_'+ str(l))

list_heads.append('blank')    
print(list_heads)
print(type(list_heads))
print(len(list_heads))
print(list_folds)
```

```python
Gaus_pd = pd.read_csv('GaussianNB1.txt', sep=" ", header=None) #, names=list_heads)
RaF_pd = pd.read_csv('RandomForest1.txt', sep=" ", header=None) #, names=list_heads)
SVC_pd = pd.read_csv('SVC1.txt', sep=" ", header=None) #, names=list_heads)
Tree_pd = pd.read_csv('DecisionTree1.txt', sep=" ", header=None) #, names=list_heads)

Gaus_pd.columns = list_heads
RaF_pd.columns = list_heads
SVC_pd.columns = list_heads
Tree_pd.columns = list_heads

del Gaus_pd['blank']
del RaF_pd['blank']
del SVC_pd['blank']
del Tree_pd['blank']
```

```python
import numpy as np
import matplotlib.pyplot as plt
#%matplotlib inline

list_folds = range(2, 50)

list_heads =[]
for l in list_folds:
    list_heads.append('mean_'+ str(l)) 


#Gaus_pd.columns = list_heads
#RaF_pd.columns = list_heads
#SVC_pd.columns = list_heads
#Tree_pd.columns = list_heads

only_means_Gaus_pd = Gaus_pd.filter(items=list_heads)
only_means_RaF_pd = RaF_pd.filter(items=list_heads)
only_means_SVC_pd = SVC_pd.filter(items=list_heads)
only_means_Tree_pd = Tree_pd.filter(items=list_heads)


print(type(only_means_RaF_pd))
#print((only_means_pd))

print("list folds: ", len(list_folds))
print("list folds: ", len(list_folds[:-1]))
print("only_means_pd: ", only_means_Gaus_pd.shape)

label_exp =[]

means_Gaus_pd = []
means_RaF_pd = []
means_SVC_pd = []
means_Tree_pd = []

max_Gaus_pd = []
max_RaF_pd = []
max_SVC_pd = []
max_Tree_pd = []

std_Gaus_pd = []
std_RaF_pd = []
std_SVC_pd = []
std_Tree_pd = []




for l in range(len(list_folds)):
    #only_means_pd.loc[l].plot( alpha=0.3) #, label="a")
    #print("exp: ", RaF_pd["fold_2"][l], " mean: ", round(np.mean(only_means_pd.loc[l]), 3), " max: ",  round(np.max(only_means_pd.loc[l]), 3))
    label_exp.append(RaF_pd["fold_2"][l])
    
    means_Gaus_pd.append(round(np.mean(only_means_Gaus_pd.loc[l]), 3))
    means_RaF_pd.append(round(np.mean(only_means_RaF_pd.loc[l]), 3))
    means_SVC_pd.append(round(np.mean(only_means_SVC_pd.loc[l]), 3))
    means_Tree_pd.append(round(np.mean(only_means_Tree_pd.loc[l]), 3))

    std_Gaus_pd.append(round(np.std(only_means_Gaus_pd.loc[l]), 3))
    std_RaF_pd.append(round(np.std(only_means_RaF_pd.loc[l]), 3))
    std_SVC_pd.append(round(np.std(only_means_SVC_pd.loc[l]), 3))
    std_Tree_pd.append(round(np.std(only_means_Tree_pd.loc[l]), 3))

    max_Gaus_pd.append(round(np.max(only_means_Gaus_pd.loc[l]), 3))
    max_RaF_pd.append(round(np.max(only_means_RaF_pd.loc[l]), 3)) 
    max_SVC_pd.append(round(np.max(only_means_SVC_pd.loc[l]), 3)) 
    max_Tree_pd.append(round(np.max(only_means_Tree_pd.loc[l]), 3))
    


type(np.mean(only_means_Gaus_pd.loc[0]))
```

```python
RaF_pd.plot(kind="scatter", x="mean_2", y="mean_3", alpha=0.4, 
             label="mean_2", c="mean_4", cmap=plt.get_cmap("jet"), colorbar=True,
            )
plt.legend()
```

# Mean performance at different houg configurations

```python
#label_exp 
#means_Gaus_pd , means_RaF_pd , means_SVC_pd , means_Tree_pd 
#max_Gaus_pd  ,max_RaF_pd  , max_SVC_pd , max_Tree_pd 
#marker='o' ,  marker='^'



plt.figure(figsize=(12,5))
print(len(label_exp))
plt.errorbar(x=label_exp, y= means_Gaus_pd, yerr=std_Gaus_pd, c= "red",markersize= 10, alpha=0.6,fmt='o', 
             markeredgecolor="black" ,  label="Gauss")
plt.errorbar(label_exp, means_RaF_pd,  yerr=std_RaF_pd, c= "green", alpha=0.6, markersize= 10, fmt='o', 
             markeredgecolor="black" ,label="RaF")
#plt.errorbar(label_exp, means_SVC_pd,  yerr=std_SVC_pd, c= "blue", alpha=0.6,  fmt='o', label="SVM")
plt.errorbar(label_exp, means_Tree_pd, yerr=std_Tree_pd, c= "yellow", alpha=0.6, markersize= 10, fmt='o', 
             markeredgecolor="black" ,label="DT")


plt.legend(['Gauss', 'RaF',  "DT"])
locs, labels = plt.xticks()
plt.setp(labels, rotation=45)
#plt.xlim([0,54])


plt.ylim([0.4, 1.2])
plt.tight_layout()

plt.xlabel("Experiments with different Hough parameters")
plt.ylabel("Accuracy ")
plt.title("Mean Performance")
plt.grid(alpha=0.2)
plt.savefig("mean_result.pdf")
plt.show()
```

#  Maximum performance at different houg configurations

```python
plt.figure(figsize=(12,5))
print(len(label_exp))
plt.errorbar(label_exp, max_Gaus_pd, yerr=std_Gaus_pd, c= "red",markersize= 10, alpha=0.6,fmt='^',
             markeredgecolor="black" ,  label="Gauss")
plt.errorbar(label_exp, max_RaF_pd,  yerr=std_RaF_pd, c= "green", alpha=0.6, markersize= 10, fmt='^',
             markeredgecolor="black" ,  label="Gauss")
#plt.errorbar(label_exp, means_SVC_pd,  yerr=std_SVC_pd, c= "blue", alpha=0.6,  fmt='o', label="SVM")
plt.errorbar(label_exp, max_Tree_pd, yerr=std_Tree_pd, c= "yellow", alpha=0.6, markersize= 10, fmt='^',
             markeredgecolor="black" ,  label="Gauss")


plt.legend(['Gauss', 'RaF',  "DT"])
locs, labels = plt.xticks()
plt.setp(labels, rotation=45)
#plt.xlim([0,54])


plt.ylim([0.4, 1.2])
plt.tight_layout()

plt.xlabel("Experiment with different Hough parameters")
plt.ylabel("Accuracy ")
plt.title("Maximum Performance")
plt.grid(alpha=0.2)
plt.savefig("max_result.pdf")
plt.show()
```
